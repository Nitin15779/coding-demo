```python
import matplotlib.pyplot as plt
overs = [0, 10, 20, 30, 40, 50]
runs = [0, 50, 90, 170, 225, 295]
chasing_runs = [0, 60, 120, 190, 300]

plt.plot(overs, runs, marker= 'o', color='green', label='Team 1')
plt.plot(overs[:len(chasing_runs)], chasing_runs, marker= 'o', color='red', label='Team 2')

plt.xlabel("Overs")
plt.ylabel("Runs")
plt.title("Batting Summary")
plt.legend()

plt.show()

```
